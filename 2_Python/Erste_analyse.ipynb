<h1>Kundendemografie</h1>
<h2>Altersverteilung</h2>

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 1, 1)
plt.hist(df['Age'], bins=20, color='#6bbf59', edgecolor='black', alpha=0.8)

plt.title('Altersverteilung', fontsize=16, fontweight='bold')
plt.xlabel('Alter', fontsize=12)
plt.ylabel('H√§ufigkeit', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.5)  # dezente horizontale Gitterlinien
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.subplot(1, 1, 1)

gender_counts = df['Gender'].value_counts()
farben = ['#4db6ac', '#81c784']

plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=farben,
    wedgeprops={'edgecolor': 'black'})
plt.title('Geschlechterverteilung', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

city_counts = df['City'].value_counts().head(10)

plt.barh(city_counts.index, city_counts.values, color='green', alpha = 0.8)

plt.title('Top 10 St√§dte', fontsize=14, fontweight='bold')
plt.xlabel('Anzahl der Kunden', fontsize=12)
plt.gca().invert_yaxis()   # oben = gr√∂√üter Wert ‚Üí sch√∂ner f√ºrs Auge

plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

<h1>Time Series Analysis</h1>

```python
df['Jahre_Monat'] = df['Date'].dt.to_period('M')

monthly_sales = df.groupby('Jahre_Monat').agg({
    'Total_Amount': 'sum',
    'Order_ID': 'count'
}).reset_index()
monthly_sales['Jahre_Monat'] = monthly_sales['Jahre_Monat'].astype(str)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 6))

axes[0].plot(monthly_sales['Jahre_Monat'], monthly_sales['Total_Amount'], 
             marker='o', linewidth=3, color='blue')
axes[0].set_title('Monatliche Umsatzentwicklung', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Umsatz (Mio. TRY)')
axes[0].get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, pos: f'{x/1e6:.1f}M')
)
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True, alpha=0.3)


axes[1].plot(monthly_sales['Jahre_Monat'], monthly_sales['Order_ID'], 
             marker='s', linewidth=3, color='green')
axes[1].set_title('Monatliches Bestellvolumen', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Monat')
axes[1].set_ylabel('Anzahl der Bestellungen')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
